```python
import pandas as pd
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import time
```

Using our cleaned functions in an initial attempt to scrape Baseball Reference.

```python
# Grabbing the relevant functions from their containing file

from ipynb.fs.full.scrape_functions import yearly_links, team_data
```

```python
# Iterating over the full collection of years

for year in range(1998, 2020):
    
    links = yearly_links(year, teams_excl=[7]) # excluding foreign rookie ball
    
    time.sleep(2)
    
    # For each org in list -> for each level in org -> for each team in level
    
    for org in links:
        for level in org:
            for link in level:
                
                team = team_data(link)
                
                if len(team) != 1: # i.e., if the link exists
                    
                    lev = team[0].replace('&nbsp;', ' ') # replaces whitespace
                    year = team[1]
                    org = team[2]
                    bat = team[3]
                    pitch = team[4]
                    
                    # Converts dataframes to CSVs and saves in proper location
                    # Accounts for possibility of an org having multiple teams at a level
                    
                    bat.to_csv(f'../Data/Raw/bat/{year}/{lev}/{year}_{lev}_{org}_bat_{level.index(link)}.csv', index=False) # bat
                    pitch.to_csv(f'../Data/Raw/pitch/{year}/{lev}/{year}_{lev}_{org}_pitch_{level.index(link)}.csv', index=False) # pitch
                    
                time.sleep(2)
```
